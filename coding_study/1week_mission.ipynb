- [í•„ìˆ˜ ì¡°ê±´] groupby ë‚˜ pivot_tableì„ í™œìš©í•©ë‹ˆë‹¤. groupby ë¡œ êµ¬í•  ë•Œ unstack()ì´ë¼ëŠ” ê¸°ëŠ¥ì„ ì‚¬ìš©í•˜ì—¬ ì•„ë˜ì™€ ê°™ì´ ì»¬ëŸ¼ì— ì¸ë±ìŠ¤ ê°’ì„ ì˜¬ë ¤ì„œ í‘œê¸°í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
- ë§Œì•½ í–‰ì—ëŠ” 'ì—°ë„'ê°€, ì—´ì— 'êµ­ê°€'ê°€ ë“¤ì–´ìˆê³  ìˆ˜ì¹˜ ë°ì´í„°ì˜ ê²°ê³¼ê°’ì´ ì•„ë˜ì™€ ê°™ë‹¤ë©´, ì¶œë ¥í˜•íƒœëŠ” ì¡°ê¸ˆ ë‹¬ë¼ë„ ê´œì°®ìŠµë‹ˆë‹¤.
- ë°ì´í„°ëŠ” ë‹¤ìŒì˜ ë°©ë²•ìœ¼ë¡œ ì½ì–´ì˜µë‹ˆë‹¤.

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
df
```

```python
df['Country'].value_counts()
df['Year'].value_counts()
```

```python
# 2011ë…„ ì´ìƒì˜ ê°’ì„ ê°€ì§„ í–‰ë§Œ ê°€ì ¸ì˜¤ê¸°
df = df[df['Year'] >= 2011]
df
```

```python
# 'Year'ë¥¼ ë½‘ì•„ì„œ ì¤‘ë³µì œê±°í•œ ê°’ì„ new_index ë¦¬ìŠ¤íŠ¸ë¡œ ìƒì„±í•œë‹¤.
index = df['Year'].values

new_index = []

for i in index:
    if i not in new_index:
        new_index.append(i)

new_index
```

```python
# 'Country'ë¥¼ ë½‘ì•„ì„œ ì¤‘ë³µì œê±°í•œ ê°’ì„ new_columns ë¦¬ìŠ¤íŠ¸ë¡œ ìƒì„±í•œë‹¤.
columns = df['Country'].values

new_columns = []

for i in columns:
    if i not in new_columns:
        new_columns.append(i)

new_columns
```

```python
# 'Life_Expectancy'ë¥¼ ë½‘ì•„ì„œ reshapeë¥¼ í™œìš©í•´ì„œ new_values ë¦¬ìŠ¤íŠ¸ë¡œ ìƒì„±í•œë‹¤.
values = df['Life_Expectancy'].values

new_values = values.reshape(10,6)
new_values
```

```python
result = pd.DataFrame(new_values, index=new_index, columns=new_columns)

result
```
