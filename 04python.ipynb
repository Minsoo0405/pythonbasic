```python
price = 10
number = int(input(“발주 수량을 입력하세요.“))
print(“발주 수량을 입력하세요.“, number)
if number < 1000:
    print(number, “개를 주문하셨습니다.“)
elif number < 1500:
    discount = 0.05
    print(number, “개를 주문하셨습니다.“)                                       # 반복되는 문구이므로 elif문에 계속적으로 작성하지 않아도 됨.
    print(“적용 할인율은“, discount * 100, “%입니다.“)                          # 반복되는 문구이므로 elif문에 계속적으로 작성하지 않아도 됨.
    print(“총 발주 금액은 $“,int(number * price * (1 - discount)), “입니다.“)   # 반복되는 문구이므로 elif문에 계속적으로 작성하지 않아도 됨.
elif number < 2000:
    discount = 0.10
    print(number, “개를 주문하셨습니다.“)
    print(“적용 할인율은“, discount * 100, “%입니다.“)
    print(“총 발주 금액은 $“,int(number * price * (1 - discount)), “입니다.“)
else:
    discount = 0.2
    print(number, “개를 주문하셨습니다.“)
    print(“적용 할인율은“, discount * 100, “%입니다.“)
    print(“총 발주 금액은 $“,int(number * price * (1 - discount)), “입니다.“)
```

```python
price = 10
quant = int(input("발주 수량을 입력하세요."))                                   # 변수 지정을 위해 quat는 if문 시작 전에 입력.

if quant < 1000:
    discount = 0
elif quant < 1500:
    discount = 0.05
elif quant < 2000:
    discount = 0.10
else:
    discount = 0.2

print("발주 수량을 입력하세요.", quant)                                         # 결과값을 나타내는 구문이고 반복되므로 마지막에 작성
print(quant, "개를 주문하셨습니다.")
print("적용 할인율은", discount * 100, "%입니다.")
print("총 발주 금액은 $",int(quant * price * (1 - discount)), "입니다.")


```
