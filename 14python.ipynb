### DataFrame apply() 메서드

    1. 특징 
        1.1. 첫 인자로 함수를 필수 값으로 받음. / 경우에 따라 두 번째 인자로 axis를 사용할 수있음(default = 0)
            1.1.1. axis가 0 or 'index'인 경우 각 column에 대해 함수를 적용
            1.1.2. axis가 1 or 'columns'인 경우 각 row에 대해 함수를 적용

### DataFrame fillna() 메서드

    1. 특징
        1.1. NaN 값을 원하는 값으로 바꿀 수 있음. / 첫 인자로 NaN을 변경하고자 하는 값을 전달하면 됨.

### DataFrame astype() 메서드

    1. 특징
        1.1. column의 자료형을 바꾸는 것

### DataFrame 실수 값을 카테고리 값으로 변환

    1. cut : 실수 값의 경계선을 지정하는 경우
        - x = 1차원 형태의 배열 형태가 옵니다.
        - bins = int, 스칼라를 요소로 갖는 시퀀스가 옵니다.
    2. qcut: 개수가 똑간은 구간으로 나누는 경우(분위수)
        - x = 1d ndarray 혹은 Series
        - q = int 혹은 분위수를 나타내는 1이하의 실수를 요소로 갖는 list (ex.[0, 0.25, 0.5, 0.75, 1.])

### DataFrame 인덱스 설정 및 제거 

    1. set_index : 기존의 row 인덱스를 제거하고 데이터 column 중 하나를 인덱스로 설정합니다.
        - 특정한 column을 인덱스로 설정할 수 있음. 이때 기존의 인덱스는 없어짐. 
    2. reset_index : 기존의 row 인덱스를 제거하고 인덱스를 데이터 열로 추가합니다.

### DataFrame 다중 인덱스

    1. row이나 column에 여러 계층을 가지는 인덱스 즉, 다중 인덱스를 설정할수도 있습니다.

### DataFrame 합성

    1. join()
    2. merge()
        - 두 데이터 프레임의 공통 column 혹은 인덱스를 기준으로 두개의 테이블을 합친다. 이 때 기준이 되는 column, row의 데이터를 키(key)라고 합니다.
        - 기본적으로는 양쪽 DataFrame에 모두 키가 존재하는 데이터만 보여주는 inner join 방식을 사용합니다.
        - outer join 방식은 키 값이 한쪽에만 있어도 데이터를 보여줍니다. (how='outer 라는 인수를 입력하면 outer join 방식으로 출력)
        - left는 첫 번째 인수 기준으로, right는 두번째 인수 기준으로 DataFrame의 키 값을 모두 보여줍니다.

```python
# DataFrame 실수 값을 카테고리 값으로 변환 - cut 
ages = [0, 2, 10, 21, 23, 37, 31, 61, 20, 41, 32, 101]
bins = [1, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년", "장년", "중년", "노년"]
cats = pd.cut(ages, bins, labels=labels)    
cats        # ages[0] = NaN, ages[1] = '미성년자', ... , ages[-2] = '장년', ages[-1] = NaN 
# cut 명령을 사용하면 실수값을 다음처럼 카테고리 값으로 바꿀 수 있습니다. bins 인수는 카테고리를 나누는 기준값이 됩니다. 영역을 넘는 값은 NaN으로 처리됩니다.
```

```python
cats.categories     # '미성년자' = 0, '청년' = 1, ... , '노년' = 4
```

```python
cats.codes          # '미성년자' = 0, '청년' = 1, ... , '노년' = 4
```

```python
# DataFrame 다중 인덱스
# - stack() 메서드 : column 인덱스 -> row 인덱스로 변환 / default: level = -1(90도 회전한 형태로 row로 변환)
# - unstack() 메서드 : row 인덱스 -> column 인덱스로 변환 / default : level = -1
# 인덱스를 지정할 때는 문자열 이름과 순서를 표시하는 숫자 인덱스를 모두 사용할 수 있습니다.
# 인덱스 값이 하나의 label이나 숫자가 아니라 ()로 둘러싸인 튜플이 되어야 합니다.
# loc 인덱서를 사용하는 경우에도 마찬가지로 튜플을 사용해서 인덱싱해야합니다.
# iloc 인덱서를 사용하는 경우에는 튜플 형태의 다중인덱스를 사용할 수 없습니다.
# 하나의 레벨 값만 넣으면 다중 인덱스중 가장 상위의 값을 지정한 것으로 반환합니다.
# 특정 레벨의 모든 인덱스 값을 인덱싱할 때는 슬라이스를 사용합니다. 인덱스의 튜플내에서는 콜론 즉 슬라이스 기호를 사용할 수 없고, slice()를 대신해서 사용.
# 다중 인덱스의 인덱스 순서를 바꾸고 싶은면 swaplevel(i, j, axis) / axis = 0일때 row, axis = 1일때 column(default는 행)
```
