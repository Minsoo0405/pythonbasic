### 1.10 데이터 색인하기

#### 1.10.1 서울 내에 있는 종합병원을 구별로 나타내기

```python
# '상권업종소분류명'이 종합병원인 데이터만 가져오기
df_hospital = df[df['상권업종소분류명'] == '종합병원']
df_hospital['상권업종소분류명'].value_counts()
```

```python
# 서울에 있는 종합병원인 데이터만 가져오기
df_seoul_hospital = df[(df['상권업종소분류명'] == '종합병원') & (df['시도명'] == '서울특별시')]
df_seoul_hospital.head(3)
```

```python
# 서울에 있는 종합병원의 구별 개수
df_seoul_hospital['시군구명'].value_counts().head(5)
```

```python
# 서울의 구별로 종합병원의 백분율
df_seoul_hospital['시군구명'].value_counts(normalize=True).head(5)
```

```python
# 상호명에 '종합병원'이라는 문구가 들어가지 않은 데이터만 가져오기
df_seoul_hospital.loc[~df_seoul_hospital['상호명'].str.contains('종합병원'), '상호명'].unique()
```

```python
# 제거할 데이터 지정하기("꽃배달|의료기|장례식장|상담소|어린이집")
drop_rows = df_seoul_hospital[df_seoul_hospital['상호명'].str.contains('꽃배달|의료기|장례식장|상담소|어린이집')].index
drop_rows = drop_rows.tolist()
drop_rows
```

```python
# 제거할 데이터 지정하기("의원")
drop_rows2 = df_seoul_hospital[df_seoul_hospital['상호명'].str.endswith('의원')].index
drop_rows2 = drop_rows2.tolist()
drop_rows2
```

```python
drop_rows = drop_rows + drop_rows2
len(drop_rows)
```

```python
# 데이터 제거하기
print(df_seoul_hospital.shape)
df_seoul_hospital = df_seoul_hospital.drop(drop_rows, axis=0)
print(df_seoul_hospital.shape)
```

```python
df_seoul_hospital['시군구명'].value_counts().plot.bar()
```

```python
plt.figure(figsize=(15,6))
sns.countplot(data=df_seoul_hospital, x='시군구명', order=df_seoul_hospital['시군구명'].value_counts().index)
```

### 1.11 folium 으로 위경도와 주소 데이터를 활용해 지도에 표현하기

#### 1.11.1 folium 설치 및 라이브러리 임포트

```python
!pip install folium
```

```python
import folium
```

```python
# 지도의 중심을 지정하기 위해 위도와 경도의 평균을 구하여 지도 나타내기
map = folium.Map(location=[df_seoul_hospital['위도'].mean(), df_seoul_hospital['경도'].mean()], zoom_start=12)
```

```python
# 마커와 팝업을 나타내기 위한 설정하기
for i in df_seoul_hospital.index:
    name = df_seoul_hospital.loc[i, '상호명']
    address = df_seoul_hospital.loc[i, '도로명주소']
    popup = f"{name} - {address}"
    location = [df_seoul_hospital.loc[i, '위도'], df_seoul_hospital.loc[i, '경도']]
    
    folium.Marker(location=location, popup=popup).add_to(map)

map
```
