### 1.5 결측치 확인하기

```python
null_count = df.isnull().sum()
null_count
```

```python
# 결측치를 막대그래프로 표현
null_count.plot.barh(figsize=(5,7))
```

### 1.6 컬럼명 변경 & 정렬하기

```python
df_null_count = null_count.reset_index()
df_null_count
```

```python
# 컬럼명 변경
df_null_count.columns = ['컬럼명', '결측치 수']
```

```python
# 결측치 많은 순으로 내림차순 정렬한 후, 상위 10개만 뽑기
df_null_count_top = df_null_count.sort_values(by='결측치 수', ascending=False).head(10)
df_null_count_top
```

### 1.7 특정 값만 가져오기

```python
drop_columns = df_null_count_top['컬럼명'].to_list()
drop_columns
```

```python
df = df.drop(drop_columns, axis=1)
```

```python
df.info()
```
